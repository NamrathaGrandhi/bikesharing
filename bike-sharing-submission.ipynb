## dropping the weathersit column as this is converted to a categorical variable 

```python
daily_data.drop(['weathersit','casual','registered','dteday','instant' ],axis=1,inplace=True)
```

## extracting dummy data for seaon variable 

```python
daily_data['seasons'] = daily_data['season'].apply(lambda x : 'spring' if x == 1 else ('summer' if x == 2 else ('fall' if x == 3 else 'winter')))
```

```python
seasons = pd.get_dummies(daily_data['seasons'])
```

## now we drop the column fall and seasons

```python
seasons = pd.get_dummies(daily_data['seasons'],drop_first=True)
daily_data = pd.concat([daily_data,seasons],axis=1)
```

```python
daily_data.drop(['season','seasons'],axis=1,inplace=True)
```

Adding methods for building models and VIF calculation

```python
def variance_calculation(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'],2)
    vif = vif.sort_values(by='VIF',ascending=False)
    return vif
```

```python
import statsmodels.api as sm  

# X_train_daily_lm = sm.add_constant(X_train_daily)


linear_regression_model1 =  build_model(X_train_daily,y=y_train_daily)               #sm.OLS(y_train_daily,X_train_daily_lm).fit()

linear_regression_model1.summary()
```

### we could see that there is a high p-value we drop the temp feature and build a new model

```python
# X_train_daily_lm = sm.add_constant(X)
X = X_train_daily.drop('temp',axis=1)
linear_regression_model2 = build_model(X,y=y_train_daily) # sm.OLS(y_train_daily ,X_train_daily_lm).fit()
linear_regression_model2.summary()
```

```python
vif_2 = variance_calculation(X)
vif_2
```

### we can see that hum has high VIF , so we drop hum and rebuild the module

```python
X = X.drop('hum',axis=1)
# X_train_daily_lm = sm.add_constant(X)

linear_regression_model3 = build_model(X,y=y_train_daily) #sm.OLS(y_train_daily , X_train_daily_lm).fit()
linear_regression_model3.summary()
```

```python
vif_3 = variance_calculation(X)
vif_3
```

### dropping month and building a new model because of its high VIF

```python
X = X.drop('mnth',axis=1)
linear_regression_model4 =  build_model(X,y=y_train_daily)    #sm.OLS(y_train_daily , X_train_daily_lm).fit()
linear_regression_model4.summary()
```

```python
vif_4 = variance_calculation(X)
vif_4
```

### as there is high VIF for windspeed dropping the feature


```python
X = X.drop('windspeed',axis = 1)
linear_regression_model5 = build_model(X,y=y_train_daily)
linear_regression_model5.summary()
```

```python
vif_5 = variance_calculation(X)
vif_5
```

### the model 5 looks promising with a better R-sqaured and adjusted R-squared , and the VIF test for correlation shows all the features have less correlation or < 5 but we see that for workingday there is high P-value so we will drop this feature a build a new model and see if there is any significant difference

```python
# summer
x_extra_test = X.drop('summer',axis=1)
linear_regression_model6 = build_model(X_features=x_extra_test , y=y_train_daily)
linear_regression_model6.summary()
```

```python
vif_6 = variance_calculation(x_extra_test)
vif_6
```

```python
x_extra_test = x_extra_test.drop('workingday',axis=1)
linear_regression_model7 = build_model(X_features=x_extra_test , y=y_train_daily)
linear_regression_model7.summary()
```

```python
vif_7 = variance_calculation(x_extra_test)
vif_7
```

## The final model linear model 7 makes more sense with all the features falling with in the expected P-value range and with better R-squared and R-adjusted 

### residula analysis on the train data by using the new trained model

```python

X_final_pred = sm.add_constant(x_extra_test)

y_train_pred = linear_regression_model7.predict(X_final_pred)
```
